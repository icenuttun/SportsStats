```python
import numpy as np
import pandas as pd
import math
```

```python
athlete_events=pd.read_csv('athlete_events.csv')
```

```python
noc_regions=pd.read_csv('noc_regions.csv')
```

```python
athlete_events.info()
```

```python
avgbysex=athlete_events.groupby('Sex').agg(mean_h=('Height','mean'),mean_w=('Weight','mean'))
print(avgbysex)
```

```python
mean_h_F=avgbysex.loc['F','mean_h']
mean_h_M=avgbysex.loc['M','mean_h']
mean_w_F=avgbysex.loc['F','mean_w']
mean_w_M=avgbysex.loc['M','mean_w']
```

```python
athlete_events.iloc[2,4]
```

```python
athlete_events2=athlete_events.copy()
```

```python
for i in range (0,athlete_events2.shape[0]):
    if athlete_events2.iloc[i,2]=='F' :
        if math.isnan(athlete_events2.iloc[i,4]) == True :
            athlete_events2.iloc[i,4]=mean_h_F
        if math.isnan(athlete_events2.iloc[i,5]) == True :
            athlete_events2.iloc[i,5]=mean_w_F
    elif  athlete_events2.iloc[i,2]=='M':
        if math.isnan(athlete_events2.iloc[i,4]) == True :
            athlete_events2.iloc[i,4]=mean_h_M
        if math.isnan(athlete_events2.iloc[i,5]) == True :
            athlete_events2.iloc[i,5]=mean_w_M
```

```python
athlete_events2
```

```python
athlete_events2['Medal'].fillna("none", inplace = True)
```

```python
athlete_events2
```

```python
athlete_events2.describe()
```

```python
dummies1 = pd.get_dummies(athlete_events2['Medal']).rename(columns=lambda x: 'Medal_' + str(x))
athlete_events_dummies = pd.concat([athlete_events2, dummies1], axis=1)
```

```python
athlete_events_dummies.drop(['Medal'], inplace=True, axis=1)
```

```python
athlete_events_dummies
```

```python
dummies2 = pd.get_dummies(athlete_events_dummies['Sex']).rename(columns=lambda x: 'Sex_' + str(x))
athlete_events_dummies = pd.concat([athlete_events_dummies, dummies2], axis=1)
```

```python
athlete_events_dummies.drop(['Sex'], inplace=True, axis=1)
```

```python
athlete_events_dummies
```

```python
athlete_events_dummies.describe()
```

```python
from pandasql import sqldf 
pysqldf=lambda q: sqldf(q,globals())
```

```python
athlete_gold=pysqldf ('select * from athlete_events_dummies where Medal_Gold=1')
athlete_silver=pysqldf ('select * from athlete_events_dummies where Medal_Silver=1')
athlete_bronze=pysqldf ('select * from athlete_events_dummies where Medal_Bronze=1')
athlete_none=pysqldf ('select * from athlete_events_dummies where Medal_none=1')
```
